```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
iris_df = sns.load_dataset('iris')
iris_df
```

```python
iris_df['species'].value_counts()
```

```python
sns.scatterplot(iris_df['sepal_length'],iris_df['petal_length'],color = 'red')
plt.show()
```

```python
sns.jointplot(iris_df['sepal_length'],iris_df['sepal_width'],kind = 'kde')
plt.show()
```

```python
sns.jointplot(iris_df['sepal_length'],iris_df['sepal_width'],kind = 'hex')
plt.show()
```

```python
sns.jointplot(iris_df['sepal_length'],iris_df['sepal_width'],kind = 'hex',color = 'red')
plt.show()
```

```python
sns.countplot(iris_df['species'])
plt.show()
```

```python
sns.countplot(y = iris_df['species'])
plt.show()
```

```python
sns.countplot(iris_df['species'],palette = sns.cubehelix_palette())
plt.show()
```
